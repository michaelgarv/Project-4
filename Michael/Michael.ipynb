```python
import pandas as pd
```

```python
batting_path = "../Data/Batting.csv"
team_path = "../Data/Teams.csv"
```

```python
batting_df = pd.read_csv(batting_path)
team_df = pd.read_csv(team_path)
```

```python
batting_df = batting_df[batting_df['yearID'] >= 2017]
grouped_batting_df = batting_df.groupby(['teamID','yearID']).agg({'G': 'sum',
    'AB': 'sum',
    'R': 'sum',
    'H': 'sum'})
```

```python
group_batting_df_reset = grouped_batting_df.reset_index()
```

```python
team_df = team_df[team_df['yearID'] >= 2015]
grouped_team_df = team_df.groupby(['teamID','yearID']).agg({'G': 'sum',
    'G': 'sum',
    'W': 'sum',
    'L': 'sum',
    'H': 'sum',
    'AB': 'sum',
    'SB': 'sum',
    'HR':'sum',
    '2B': 'sum',
    '3B': 'sum',
    'R':'sum',
    'ERA': 'sum',
    'RA': 'sum',
    'IPouts':'sum',
    'HA': 'sum',                                                       
    'BBA': 'sum',
      })

grouped_team_df['BAVG'] = grouped_team_df['H'] / grouped_team_df['AB']
grouped_team_df['SLG'] = (grouped_team_df['H'] - grouped_team_df['2B'] - grouped_team_df['3B'] - grouped_team_df['HR'] +  2*grouped_team_df['2B'] + 3*grouped_team_df['3B'] + 4*grouped_team_df['HR']) / grouped_team_df['AB']
grouped_team_df['H/BBA'] = grouped_team_df['HA']+grouped_team_df['BBA'] 
grouped_team_df['IP'] = grouped_team_df['IPouts']/3 
grouped_team_df['WHIP'] = grouped_team_df['H/BBA']/grouped_team_df['IP'] 
grouped_team_df.head()
```

```python
grouped_team_df_reset = grouped_team_df.reset_index()
grouped_team_df_reset = grouped_team_df_reset[grouped_team_df_reset['yearID'] != 2020]
#grouped_team_df_reset = grouped_team_df_reset[grouped_team_df_reset['yearID'] != 2022]


testing_df = grouped_team_df_reset[grouped_team_df_reset['yearID'] == 2022]
grouped_team_df_reset.head(7)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Extract features and target variable
X = grouped_team_df_reset[['WHIP','IPouts']].values
y = grouped_team_df_reset['RA'].values

# Split the data into training and testing sets (80% for training, 20% for testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```
